```python
import pandas as pd
from pathlib import Path
```

```python
df = pd.read_csv(Path("../data/turnover.csv"))
```

```python
df.isna().mean().sort_values(ascending = False).head(5)
```

```python
df = df.dropna(subset='Idade')
```

```python
df.isna().mean().sort_values(ascending = False).head(5)
```

```python
len(df)
```

```python
df['RendaMensal'].hist()
```

```python
media = df['RendaMensal'].mean()
dp = df['RendaMensal'].std()
ls = media + 2*dp
li = media - 2*dp
df = df[(df['RendaMensal'] > li) & (df['RendaMensal'] < ls)]
```
